```python
df["Postal Code"].fillna(df["Postal Code"].mean(),inplace=True)
```

```python
top_customers = df.groupby('Customer Name').sum().sort_values('Sales',ascending=False).head(25)
top_customers = top_customers[["Sales"]].round(2)
top_customers.reset_index(inplace=True)


```

```python
plt.figure(figsize=(10,10))
plt.title("Top 25 Customers")
plt.bar(top_customers['Customer Name'],top_customers['Sales'],color='magenta')
plt.xlabel("Customer Name")
plt.ylabel("Sales")
plt.xticks(rotation=90)
plt.show()
```

```python
#To show more data throught plots
plt.figure(figsize=(10,10))
plt.title("Top 25 Customers")
plt.bar(top_customers['Customer Name'],top_customers['Sales'],color='mediumseagreen')
plt.xlabel("Customer Name")
plt.ylabel("Sales")
plt.xticks(rotation=90)

for i, sales_value in enumerate(top_customers['Sales']):
   plt.text(i, sales_value-8000, str(sales_value), fontsize=12,rotation=90,color='k', horizontalalignment='center')
plt.show()
```

```python
top_regions = df.groupby('Region').sum().sort_values('Sales',ascending=False).head(25)
top_regions = top_regions[["Sales"]].round(2)
top_regions.reset_index(inplace=True)
```

```python
plt.figure(figsize=(10,10))
plt.title("Top 25 Regions")
plt.bar(top_regions['Region'],top_regions['Sales'],color='yellowgreen')
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)

for i, sales_value in enumerate(top_regions['Sales']):
   plt.text(i, sales_value-8000, str(sales_value), fontsize=12,rotation=90,color='k', horizontalalignment='center')
plt.show()
```

```python
top_states = df.groupby('State').sum().sort_values('Sales',ascending=False).head(25)
top_states = top_states[["Sales"]].round(2)
top_states.reset_index(inplace=True)
```

```python
plt.figure(figsize=(10,10))
plt.title("Top 25 States")
plt.bar(top_states['State'],top_states['Sales'],color='yellowgreen')
plt.xlabel("States")
plt.ylabel("Sales")
plt.xticks(rotation=90,fontsize=10)
plt.yticks(fontsize=10)

for i, sales_value in enumerate(top_states['Sales']):
   plt.text(i, sales_value-8000, str(sales_value), fontsize=12,rotation=90,color='k', horizontalalignment='center')
plt.show()
```

```python
state = ['Alabama', 'Arizona' ,'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware', 'Florida',
         'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland',
         'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi', 'Missouri', 'Montana','Nebraska', 'Nevada', 'New Hampshire',
         'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania',
         'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington',
         'West Virginia', 'Wisconsin','Wyoming']
state_code = ['AL','AZ','AR','CA','CO','CT','DE','FL','GA','HI','ID','IL','IN','IA','KS','KY','LA','ME','MD','MA',
              'MI','MN','MS','MO','MT','NE','NV','NH','NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN',
              'TX','UT','VT','VA','WA','WV','WI','WY']
```

```python
state_df = pd.DataFrame(state, state_code)
state_df.reset_index(level=0, inplace=True)
state_df.columns = ['State Code','State']
sales = df.groupby(["State"]).sum().sort_values("Sales", ascending=False)
sales.reset_index(level=0, inplace=True)
sales.drop('Postal Code',axis=1, inplace = True)
sales= sales.sort_values('State', ascending=True)
sales.reset_index(inplace = True)
sales.drop('index',axis = 1,inplace = True)
sales.insert(1, 'State Code', state_df['State Code'])
```

```python
import plotly.graph_objects as go

sales['text'] = sales['State']
fig = go.Figure(data=go.Choropleth(
    locations=sales['State Code'],
    text=sales['text'],
    z = sales['Sales'].astype(float),
    locationmode = 'USA-states',
    colorscale = 'Greens',
    colorbar_title = "Sales",

))

fig.update_layout(
    title_text = 'Sales',
    geo_scope='usa',
)

fig.show();
```
